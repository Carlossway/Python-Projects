## Task 1: Create a Scatter Plot
You are given sales data for five products. Visualise the relationship between the number of units sold and the revenue generated using a scatter plot.

```python
## Task 1: Create a Scatter Plot

# Import libraries
import matplotlib.pyplot as plt

# Data for scatter plot
units_sold = [10, 20, 30, 40, 50] # x-axis data
revenue = [15, 25, 35, 45, 55] # y-axis data

# Create a scatter plot
# Use plt.scatter() to plot units_sold vs. revenue. Add color='blue' and label='Revenue Data'.
# TODO: Replace the variables  and values
plt.scatter(units_sold, revenue, color='blue', label='revenue_data')


plt.xlabel('Units Sold')  #  label for the x-axis
plt.ylabel('Revenue')  # label for the y-axis

# TODO: Replace the title 
plt.title('Title of the Plot')  # Replace with the title of your plot

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
